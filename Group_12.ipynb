##Business Understanding
To understand exercise data, particularly the relationship between exercise duration, heart rate (Pulse and Maxpulse), and calories burned which in turn would be valuable in creating value in the health, fitness and wellness industries.

## Problem Statement
How to accurately predict the number of calories burned during an exercise session based on measurable workout characteristics such as duration, average pulse, and maximum pulse?


### Objectives
To analyze the relationship between exercise characteristics and calories burned. Specifically, it aims to understand:

1. What the relationship is between exercise duration and calories burned?
2. Is there a significant relationship between higher pulse rates and increased calorie burn?
3. What predictive models could provide the best accuracy for estimating calorie expenditure?
4. What the optimal combination of duration, average pulse, and max pulse for maximizing calorie burn is?

### Success criteria
1. Accuracy of the prediction model
2. Effectiveness in calories burning


### Limitations/Assumptions
1. Limited variables. Important variables such as age, gender and type of exercise are missing.
2. Lack of baseline health data to enable comparison of impact
3. Lack of temporal data to give an indication of period of time that the data was collected.
